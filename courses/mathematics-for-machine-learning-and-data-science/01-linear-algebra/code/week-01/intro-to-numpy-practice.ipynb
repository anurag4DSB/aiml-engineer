# Introduction to NumPy

Package: `numpy`

## NumPy Arrays vs Python Lists

- Python lists can hold any mix of value types; NumPy arrays can only hold one type of value.
- NumPy provides the array object, which is much faster and more compact than Python lists.
- A NumPy array object is called an `ndarray`: an "n-dimensional array".

## Importing NumPy

```python
import numpy as np
```

## Creating a 1-D Array

```python
one_d_array = np.array([10, 12])
print(one_d_array)
```

## Creating a New ndarray

1. `np.array()` — takes a list of values.
2. `np.arange(start, stop, step)` — returns an array of evenly spaced elements within a given interval (`start` defaults to 0 if only `stop` is given).
3. `np.linspace(start, stop, num=50)` — returns `num` evenly spaced samples over the interval `[start, stop]`. Default dtype is `float64`.
4. `np.zeros()` — returns an array with all values set to zero.
5. `np.ones()` — returns an array with all values set to one.
6. `np.empty()` — returns an array with uninitialized values; use only when you're about to overwrite them immediately.
7. `np.random.rand()` — returns an array with values chosen at random.

**Note:** `<U23` means a 23-character unicode string; `<` denotes little-endian byte order.
More on data types: https://numpy.org/doc/stable/user/basics.types.html

```python
basic_arr = np.array([2, 3, 4])
print(basic_arr)
```

```python
evenly_spaced_array = np.arange(4, 90, 8)
print(evenly_spaced_array)
```

```python
lin_evenly_spaced_arr = np.linspace(100, 200, 2, dtype=int)
print(lin_evenly_spaced_arr)
char_arr = np.array(['welcome to math for ml!'])
print(char_arr)
print(char_arr.dtype)
```

```python
zeros_arr = np.zeros(20)
print(zeros_arr)
```

```python
ones_arr = np.ones(2, dtype=int)
print(ones_arr)
```

```python
empty_arr = np.empty(5)
print(empty_arr)
```

```python
random_arr = np.random.rand(100)
print(random_arr)
```

## Multi-Dimensional Arrays

- Create with `np.array()` on a list of lists, e.g. `np.array([[1,2,3], [4,5,6]])`.
- Reshape with `np.reshape()` — takes a 1-D array and reshapes it into the specified dimensions `(rows, cols)`.

### Finding Size, Shape, and Dimension

- `ndim` — number of dimensions of the array.
- `shape` — shape of the array; each number in the tuple is the length of the corresponding dimension.
- `size` — total number of elements in the array.

### Math Operations

- `+`, `-`, `*` work element-wise when arrays are the same size.
- **Broadcasting** lets you operate on arrays of different shapes, e.g. multiplying a vector by a scalar.

```python
two_d_array = np.array([[1,2,3], [2,3,4]])
print(two_d_array)
reshaped = np.reshape(two_d_array, (1,6))
print(reshaped)
```

```python
one_dim_array = np.array([1,2,3,4,5,6])
reshaped_arr = np.reshape(one_dim_array, (2,3))
print(reshaped_arr)
```

```python
print(reshaped_arr.ndim)
print(reshaped_arr.size)
print(reshaped_arr.shape)
```

```python
arr_1 = np.array([5, 6, 7])
arr_2 = np.array([1, 2, 3])
print(arr_1 - arr_2)
print(arr_1 + arr_2)
print(arr_1 * arr_2)
print(arr_1 / arr_2)
```

```python
vector = np.array([1, 2, 43])
print(vector * 1.6)
print(vector - 1.6)
print(vector + 1.6)
print(vector / 1.6)
```

## Indexing and Slicing

### Indexing

- `arr[element/row]`
- `arr[row][col]`

### Slicing

- `arr[start:end:step]`
- Default `start` = 0 (inclusive)
- Default `end` = length of array
- Default `step` = 1
- Note that `a == a[:] == a[::]`

For 2-D arrays, slicing works at the row and column level — if the column isn't specified, all columns are included: `arr[1:2] == arr[1:2, :]`.

```python
print(vector[2])
print(two_d_array)
print(two_d_array[0])
print(two_d_array[0][2])
```

```python
slice_arr = np.array([1, 2, 3, 4, 5])
print(slice_arr)
print(slice_arr[1:3:4])
print(slice_arr[1:])
print(slice_arr[::2])
print(two_d_array)
print(two_d_array[0:1, :])
print(two_d_array[:,:1])
```

## Stacking

Join two or more arrays along a new axis, either horizontally or vertically.

- `np.vstack()` — stacks vertically
- `np.hstack()` — stacks horizontally
- `np.hsplit()` — splits an array into several smaller arrays

**Note:** `hsplit` operates on columns (left/right pieces); `vsplit` operates on rows (top/bottom pieces).

```python
a1 = np.array([[1, 1], [2, 2]])
a2 = np.array([[3, 3], [4, 4]])
print(f'a1:\n{a1}')
print(f'a2:\n{a2}')
```

```python
vert_stack = np.vstack((a1, a2))

print(vert_stack)
```

```python
horizontal_stack = np.hstack((a1, a2))
print(horizontal_stack)
```

```python
horizontal_split = np.hsplit(a1, 1)
print(horizontal_split)
```
